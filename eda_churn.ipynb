##Análise bivariada

```python
#Hipótese: clientes com contrato do tipo mensal são mais propensos ao churn
#Construir tabela de contigência entre tipo de contrato e churn - Quantidade

pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
#Avaliar a correlação entre duas variáveis categóricas (qualitativas)

#Executar um teste de hipótese chamado Chi_Square ou Qui-Quadrado de Pearson
#Num teste de hpótese, duas hipóteses são formuladas
#H0 (hipótese nula): as duas variáveis são independentes
#H1 (hipótese complementar): as duas variáveis não são independetes

#O teste serve para confirmar ou recusar a hipótese nula
#Quando a probabilidade de observarmos H0, é inferior a 0.05, recusamos com a hipótese nula e seguimos a complementar




```

```python
#Gerar df da crosstab (sem totais)

df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
#Calcular os scores e p-values
from scipy.stats import chi2_contingency

```

```python
#Tupla = Lista imutável

chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
#Apresentar números com decimais sem a notação científica

pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
df_chi_scores_churn_contract
```

Conforme teste estatístico (Chi_square), o p-value <=5, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação

```python
#Avaliar a correlação entre uma variável qualitativa em quantitativa
#Usar o mesmo teste Qui_square para transformar variável qualitativa em quantitativa
#Hipótese: cliente com menos de 6 meses de contrato é mais propenso ao churn

#Criar coluna nova no dataframe com base em condição de outra coluna

import numpy as np
```

```python
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6, 'Yes', 'No')
```

```python
#Gerar df da crosstab (sem totais)

df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
#Calcular os scores

chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)

```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
df_chi_scores_churn_tenure
```

Como o p-value é <= 0.05, rejeitamos o H0, ou seja, as variáveis não são independentes. E vale mencionar que com base no Qui2 resultante das 2 análises, a correlação entre Churn e TempoMenor6 meses é menor que a correlação entre Churn e Tempo de Contrato

```python
#Criar dataframe somente com clientes do tipo Month to Month

df_churn_month = df_churn[df_churn.Contract=='Month-to-month']
```

```python
df_churn_month
```

```python
#Histograma - chegar visualmente se os dados seguem uma distribuição normal

df_churn_month.TotalCharges.plot.hist()
```

```python
#Método de Tukey - IQR (Distribuição não normal)
#IQR - Range interquartil
#IQR - 3o quartil - 1o quartil

q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)

iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q1_TotalCharges_month + (iqr_TotalCharges_month * 1.5)

limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

```python
#ZScore (Usado para distribuição normal)

#Indica quantos desvios padrões um ponto específico de dados está distante da média
#zscore = (x - media) / desvio_padrão

from scipy.stats import zscore

z = np.abs(zscore(df_churn_month.TotalCharges))

z
```

```python
df_churn_month[z > 3]
```
